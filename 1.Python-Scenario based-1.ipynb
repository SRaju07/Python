```python
# 8. A banking system processes withdrawal requests and ensures the user has enough balance.
# Write logic to check if a user has enough balance before allowing a withdrawal and update the remaining balance accordingly.
```

```python
class Banking:
    def Withdraw():
        balance=float(input("Enter your balance : "))
        amount=float(input("Enter the amount to be withdraw : "))
        if(balance>=amount):
            print("You have enough balance to withdraw ",amount)
            total=balance-amount
            print("Total Balance : ",total)
        elif(balance>amount):
            print("Insufficient balance to withdraw ",amount)
      
            return balance
```

```python
Banking.Withdraw()
```

```python
# 9. A calendar system verifies whether a given year is a leap year based on standard leap year rules.
 # Write logic to determine whether a given year is a leap year.
```

```python
class LeapYear:
    def Leap():
        year=int(input("Enter the year to find leap year or not : "))
        if(year%4==0 and year%100!=0 and year%400==0):
            print(year,"Its a leap year ")
        else:
            print(year,"Its a not leap year ")
            return year
```

```python
LeapYear.Leap()
```
